```python
import pandas as pd
df = pd.read_csv("data/Storefronts_Reported_Vacant_or_Not.csv")
df.head()
```

```python
#filter the data where the reporting year is "2020-2021"
df = df[df["Reporting Year"] == "2020 and 2021"]
```

```python
df
```

```python
#filter to columns we want: Reporting Year, BOROUGH, ZIP CODE, VACANT ON 12/31, NBHD

df = df[["Reporting Year", "BOROUGH", "ZIP CODE", "VACANT ON 12/31", "NBHD"]]
```

```python
df
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
#replace in the Reporting Year column "2020 and 2021" with "2020" and then convert the column to an integer

df["Reporting Year"] = df["Reporting Year"].replace("2020 and 2021", "2020")

df["Reporting Year"] = df["Reporting Year"].astype(int)

df
```

```python
#now in the VACANT ON 12/31 column, replace the "Y" with 1 and the "N" with 0

df["VACANT ON 12/31"] = df["VACANT ON 12/31"].replace("Y", 1)

df["VACANT ON 12/31"] = df["VACANT ON 12/31"].replace("N", 0)

df

```

```python
#replace yes with 1 and no with 0 in the VACANT ON 12/31 column and convert the column to an integer

df["VACANT ON 12/31"] = df["VACANT ON 12/31"].replace("YES", 1)

df["VACANT ON 12/31"] = df["VACANT ON 12/31"].replace("NO", 0)

df["VACANT ON 12/31"] = df["VACANT ON 12/31"].astype(int)

df
```
