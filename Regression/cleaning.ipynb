In this Notebook, we will clean our data which is about whether storefront is vacant or not in New York City. The data comes from Open Nyc's website. We will process this data and make it ready for analysis.

```python
import pandas as pd
df_vacant_or_not = pd.read_csv('data_open_nyc/Storefronts_Reported_Vacant_or_Not.csv')
```

```python
df_vacant_or_not
```

Here, each row is a storefront and each column is a feature of that storefront. We will clean the data by removing the unnecessary columns, renaming the columns, and removing the rows with missing values. We will also add a new columns which will be used for analysis.

Here according to the data dictionary, Reporting Year is "End year for the end date on the primary registration and supplemental storefront registration.

For example; primary registrations that ended on December 31, 2019 and their corresponding supplemental storefront registration that ended the following year on June 30, 2020, would have "2019 and 2020" in this column."

```python
df_vacant_or_not['Reporting Year'] = df_vacant_or_not['Reporting Year'].replace('2021 and 2022', '2022')
df_vacant_or_not['Reporting Year'] = df_vacant_or_not['Reporting Year'].replace('2020 and 2021', '2021')
df_vacant_or_not['Reporting Year'] = df_vacant_or_not['Reporting Year'].replace('2019 and 2020', '2020')
```

```python
df_vacant_or_not['Reporting Year'] = df_vacant_or_not['Reporting Year'].astype(int)
```

```python
df_vacant_or_not
```

```python

df_vacant_or_not2 = df_vacant_or_not[df_vacant_or_not['Reporting Year'] == 2022]
```

```python

#now make a new df where i have columns Reporting Year, BOROUGH, VACANT ON 12/3

df_vacant_or_not3 = df_vacant_or_not2[['Reporting Year', 'BOROUGH', 'VACANT ON 12/31','ZIP CODE', 'NBHD']]
df_vacant_or_not3
```

```python
df_vacant_or_not3['VACANT ON 12/31'] = df_vacant_or_not3['VACANT ON 12/31'].replace({'YES': 1, 'NO': 0})
```

```python
df_vacant_or_not3
```

```python
#now i just want the vacant storefronts which have the value 1 in the VACANT ON 12/31 column

df2 = df_vacant_or_not3[df_vacant_or_not3['VACANT ON 12/31'] == 1]
```

```python
df2
```

```python
#now do a group by the zip code and count the number of vacant storefronts, by making a new column called total vacant storefronts, drop the 'Reporting Year' column

df3 = df2.groupby(['ZIP CODE']).count().reset_index()

df3
```

```python
#drop the columns that are not needed

df4 = df3.drop(['Reporting Year', 'VACANT ON 12/31', 'BOROUGH'], axis=1)
```

```python
df4
```

```python
#rename the column NBHD to vacant storefronts

df4.rename(columns={'NBHD': 'vacant_storefronts'}, inplace=True)
```

```python
df4
```

```python
#DROP THE FIRST ROW OF THE DATAFRAME

df4.drop(df4.index[0], inplace=True)
```

```python
df4
```

```python
df4['ZIP CODE'] = df4['ZIP CODE'].astype(int)
```

Now we can merge our census data with this data. We will merge the census data with the storefront data on the basis of zipcode.

```python
df_census = pd.read_csv('census_data-by-zipcode.csv')
```

```python
df_census.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
df_census
```

```python
#now merge the two dataframes df8 and df_census_by_zip on the zip code column. 

df_merged = pd.merge(df4, df_census, left_on='ZIP CODE', right_on='zcta', how='inner')

```

```python
df_merged
```

```python
#make a new df called df_final all the columns except zcta

df_final = df_merged.drop(['zcta'], axis=1)
```

```python
df_final
```

Here, we have the final data which is ready for analysis.
here, each row, we have the number of vacant storefronts in each zipcode. We will use this data to analyze the relationship between the number of vacant storefronts and the socio-economic features of the neighborhood.
